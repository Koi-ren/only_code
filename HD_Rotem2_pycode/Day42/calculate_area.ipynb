```python
def calculate_area(directions, lengths):
    x, y = 0, 0
    points = [(x, y)]

    for direction, length in zip(directions, lengths):
        if direction == 1:  # 동쪽
            x += length
        elif direction == 2:  # 서쪽
            x -= length
        elif direction == 3:  # 남쪽
            y -= length
        elif direction == 4:  # 북쪽
            y += length
        points.append((x, y))

    area = 0
    for i in range(len(points) - 1):
        x1, y1 = points[i]
        x2, y2 = points[i + 1]
        area += (x1 * y2 - x2 * y1)
    
    return abs(area) // 2

# 입력
soldiers_per_square_meter = int(input().strip())
directions = []
lengths = []

for _ in range(6):
    direction, length = map(int, input().strip().split())
    directions.append(direction)
    lengths.append(length)

# 면적 계산
area = calculate_area(directions, lengths)

# 총 병력 수 계산
total_soldiers = area * soldiers_per_square_meter

# 출력
print(total_soldiers)
```
