```python
INF = int(1e9) # 무한을의미하는10억설정
# 노드의개수및간선의개수입력받기
n,m = map(int,input().split())
 # 2차원리스트를만들고, 무한으로초기화
graph = [[INF]*(n+1) for _ in range(n+1)]
 # 자기자신에서자기자신으로가는비용은0으로초기화
for a in range(1, n+1):
    for b in range(1, n+1):
        if a==b:
            graph[a][b] = 0
 # 각간선에대한정보입력받아, 그값으로초기화
for _ in range(m):
    # a에서b로가는비용은c
    a, b, c = map(int, input().split())
    graph[a][b] = c
 # 점화식에따라플로이드워셜수행
for k in range(1, n+1):
    for a in range(1, n+1):
        for b in range(1, n+1):
            graph[a][b] = min(graph[a][b], graph[a][k]+ graph[k][b])
 # 수행된결과출력
for a in range(1, n+1):
    for b in range(1, n+1):
        # 도달할수없는경우, 무한으로출력
        if graph[a][b] == INF:
            print(a,"->",b,":INFINITY", end= ' ')
        # 도달할수있는경우거리를출력
        else:
            print(a,"->",b,":",graph[a][b], end=' ')
        print()
```
