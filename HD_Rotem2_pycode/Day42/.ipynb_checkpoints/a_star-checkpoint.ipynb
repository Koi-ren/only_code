```python
import heapq

# 목표 상태 정의
goal_state = (1, 2, 3, 4, 5, 6, 7, 8, 0)

# 휴리스틱 함수 (맨해튼 거리)
def heuristic(state):
    distance = 0
    for i in range(9):
        if state[i] != 0:
            goal_position = goal_state.index(state[i])
            distance += abs(goal_position // 3 - i // 3) + abs(goal_position % 3 - i % 3)
    return distance

# 가능한 이동 (위, 아래, 왼쪽, 오른쪽)
moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]

# 퍼즐 상태를 문자열로 변환 (방문 확인을 위해)
def state_to_string(state):
    return ''.join(map(str, state))

# A* 알고리즘
def a_star(start_state):
    priority_queue = []
    heapq.heappush(priority_queue, (heuristic(start_state), 0, start_state, -1, -1))
    visited = set()
    visited.add(state_to_string(start_state))
    
    while priority_queue:
        f, g, current_state, prev_row, prev_col = heapq.heappop(priority_queue)
        
        if current_state == goal_state:
            return g
        
        empty_index = current_state.index(0)
        empty_row, empty_col = empty_index // 3, empty_index % 3
        
        for move in moves:
            new_row, new_col = empty_row + move[0], empty_col + move[1]
            if 0 <= new_row < 3 and 0 <= new_col < 3:
                new_index = new_row * 3 + new_col
                new_state = list(current_state)
                new_state[empty_index], new_state[new_index] = new_state[new_index], new_state[empty_index]
                new_state = tuple(new_state)
                
                if state_to_string(new_state) not in visited:
                    visited.add(state_to_string(new_state))
                    heapq.heappush(priority_queue, (g + 1 + heuristic(new_state), g + 1, new_state, new_row, new_col))
    
    return -1

# 테스트 입력 처리
n = int(input().strip())
for _ in range(n):
    start_state = []
    for _ in range(3):
        line = input().strip().split()
        start_state.extend([int(x) if x != '#' else 0 for x in line])
    start_state = tuple(start_state)
    result = a_star(start_state)
    if result == -1:
        print("불가능")
    else:
        print(result)
```
