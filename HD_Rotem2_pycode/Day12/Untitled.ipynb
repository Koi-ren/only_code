```python
import numpy as np
import matplotlib.pyplot as plt

# 모집단 분포 (균등분포)
population = np.random.uniform(0, 100, 10000)

# 표본 크기
sample_size = 300

# 표본 평균 분포
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(1000)]

# 시각화
plt.hist(sample_means, bins=30, edgecolor='k', alpha=0.7)
plt.title('Sample Mean Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
