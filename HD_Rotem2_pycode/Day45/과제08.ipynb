```python
import random

class Player:
    def __init__(self, sname):
        self.__x = random.randrange(0, 800)
        self.__y = random.randrange(0, 600)
        self.__dis = 2
        self.__sname = sname

    @property
    def move(self):
        return f"{self.__x}, {self.__y} ===> {self.__sname}"

    @move.setter
    def move(self, bu):
        distance = self.__dis
        if bu == "Left": self.__x -= distance
        elif bu == "Right": self.__x += distance
        elif bu == "Up": self.__y += distance
        elif bu == "Down": self.__y += distance
        else: print("잘못된 입력입니다.")
        print(self.__x, ",", self.__y, "===>", self.__sname)        

p1 = Player(input("이름을 입력: "))

if __name__ == "__main__":
    print(p1.move)
    while True:
        direction = input("어디로 갈까요? (Left/Right/Up/Down): ")
        p1.move = direction
```
