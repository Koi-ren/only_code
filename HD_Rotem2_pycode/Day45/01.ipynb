```python
class Person:
    def __init__(self, fname, lname, age):
        self.first_name = fname  #Public
        self.last_name = lname
        self._age = age  #(__)언더바 두 개 추가로 privite 선언 ==> gatter, setter 메소드로 외부에서 접근 가능

    def show(self):
        print(self.first_name, ",", self.last_name, " === ", self._age)

    def get_age(self):
        return self._age

    # def set_age(self, age):
    #     if age < 0:
    #         raise ValueError("Invalid age")
    #     self._age = age

    def set_age(self, age):  #나이가 출력되게 하고 싶으면!
        if age < 0:
            print(age, ", 나이값이 유효하지 않아요~!")
        self._age = age
```

```python
p1 = Person('박', '현호', 26)
p2 = Person("Joe", "Lina", 30)
```

```python
p1.show()
p2.show()
```

```python
class Person2:
    def __init__(self, fname, lname, age):
        self.first_name = fname  #Public
        self.last_name = lname
        self._age = age  #(__)언더바 두 개 추가로 privite 선언 ==> gatter, setter 메소드로 외부에서 접근 가능

    def show(self):
        print(self.first_name, ",", self.last_name, " === ", self._age)

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, age):
        if age < 0:
            raise ValueError("Invalid age")
        self._age = age

p3 = Person2("d",'d',40)
```

```python
p4 = Person2("d","d",1)
p4.age = p4.age + 1
p4.age
```
