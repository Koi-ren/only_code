```python
class Country:
  name='국가명'
  population='인구정보'
  captial = '수도'

  def show(self):
    print(self, '국가 클래스의 메소드입니다.')


c1 = Country()
c2 = Country()

c1.show()  # ==> show(c1)
c2.show()  # ==> show(c2) <-- 다른 언어에서는 this라고 부른다.
```

```python
class Country2:
  def __init__(self, name, population, capital):
    self.name = name
    self.population = population
    self.capital = capital
  def show(self):
    # print(self.name,'국가 클래스의 메소드입니다.')
    print(self.name,"==>", self.capital)

cn1 = Country2('한국', '5천만', '서울')  ## ==> __init__(cn1, '한국','5천만','서울')로 자동으로 입력
cn2 = Country2('인도', '13억', '뉴델리')

cn1.show()
cn2.show()
```

```python
class Korea(Country2):
  def __init__(self):
    super().__init__('한국', '5천만', '서울')

k1 = Korea()
k1.show()
```
