#모듈화 = 함수 ==> 객체
모듈이란? 함수(변수)나 클래스를 독립적 단위로 원데이터에서 나아가서 사용할 수 있는 파일 단위 

모듈 명은 반드시 식별(구분)되어야한다

```python
# mod1.py
def add(a, b):
    result = 0
    for i in b:
        result += i
    return result

def sub(a, b):
    return a - b
```

```python
#add(5,3,4,3)
add(4,[5,7,9])
```

```python
# mod1.py
def add(a, *b):
    result = 0
    for i in b:
        result += i
    return result + a

def sub(a, b):
    return a - b
```

```python
add(3,5,7,9)
```
