```python
import csv

# 파일을 쓰기 모드로 열기
with open('C:/Users/plane/Desktop/weather.csv', 'w', newline='') as f:
    wr = csv.writer(f)
    
    wr.writerow(['날짜', '요일', '날씨', '최저기온', '최고기온', '강수량'])
    wr.writerow(['20230101', '월', '맑음', '-4.5', '3.8', '0'])
    wr.writerow(['20230102', '화', '맑음', '-7.4', '0.4', '0'])
    wr.writerow(['20230103', '수', '맑음', '-9.0', '0.6', '0'])
```

```python
f = open('C:/Users/plane/Desktop/weather.csv', 'r')
rdr= csv.reader(f)
#반복문으로한행씩출력하기
for line in rdr:
    print(line) 
f.close()
```
