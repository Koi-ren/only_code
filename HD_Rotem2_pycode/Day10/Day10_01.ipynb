```python
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from konlpy.tag import Okt

okt = Okt()#품사 태깅을 위한 객체를 생성

with open("C:/ws/HD_Rotem2_pycode/data_study_01/07장/대한민국헌법.txt", 'r', encoding='utf-8') as f:
    text = f.read()

hun = okt.nouns(text)

words = [n for n in hun if len(n) > 1]
c = Counter(words) #Counter 객체 만들어서 사용하기 ♪

# cloud_mask = np.array(Image.open(path.join(d, "C:/ws/HD_Rotem2_pycode/Day10/cloud.png")))
cloud_mask = np.array(Image.open("C:/ws/HD_Rotem2_pycode/Day10/cloud.png"))

wc = WordCloud(font_path='C:/Windows/Fonts/gulim.ttc',
               width=400, height=400, scale=2.0, max_font_size=250, background_color='white', mask=cloud_mask)

gen = wc.generate_from_frequencies(c) #여기서 frequency는 단어의 빈도수(주파수)를 의미한다
plt.imshow(gen) #image show
plt.axis("off")
wc.to_file('헌법_워드클라우드.png')
```

```python
from konlpy.tag import Okt
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

masking = np.array(Image.open("C:/ws/HD_Rotem2_pycode/Day10/alice.png"))

with open('C:/ws/HD_Rotem2_pycode/data_study_01/07장/대한민국헌법.txt', 'r', encoding='utf-8') as f: text = f.read()

okt = Okt()
nouns = okt.nouns(text)
words = [n for n in nouns if len(n)>1]
c = Counter(words)

wc = WordCloud(font_path='C:/Windows/Fonts/gulim.ttc', width=400, height=400, scale=2.0, max_font_size=250, background_color='white',mask = masking)
gen = wc.generate_from_frequencies(c)

plt.figure()
plt.imshow(gen)
plt.axis("off")
plt.show()
wc.to_file('헌법_워드클라우드_과제.png')
```
