```python
from sklearn.datasets import fetch_olivetti_faces

### Olivetti data는 사람의 얼굴 이미지 데이터셋으로, 사람 1명 당 10장의 얼굴 사진을 가지고 있는 데이터이다.
olivetti_faces = fetch_olivetti_faces()
print(olivetti_faces.DESCR)
```

```python
import matplotlib.pyplot as plt

olivetti_face = fetch_olivetti_faces()
```

```python
faces = olivetti_face.images[olivetti_face.target == 10]

N = 2
M = 5
fig = plt.figure(figsize=(10, 5))

for n in range(N*M):
    ax = fig.add_subplot(N, M, n+1)
    ax.imshow(faces[n], cmap=plt.cm.bone)
    ax.grid(False)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])

plt.suptitle('Faces of No.10')
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)

X = olivetti_face.data
W = pca.fit_transform(X)

X_inv = pca.inverse_transform(W) #(10, 4096) -> 4096 = 64 * 64 사진크기

N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0.5, wspace=0.05)

for n in range(N*M):
    ax = fig.add_subplot(N, M, n+1)
    ax.imshow(X_inv[n].reshape(64, 64), cmap=plt.cm.bone) # 사진 크기대로 다시 reshape

plt.suptitle('PCA result')
plt.tight_layout()
plt.show()
```
