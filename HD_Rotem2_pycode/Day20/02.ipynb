```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#def logistic(x, beta_0, beta_1): return 1 / (1 + np.exp(-(beta_0 + beta_1 * x)))
logistic = lambda x, beta0, beta1: 1 / (1 + np.exp(-(beta0 + beta1 * x)))
```

```python
beta0, beta1, x = 0, -1, np.linspace(-10,10)
```

```python
plt.plot(x,logistic(x,beta0,beta1))
plt.show()
```
