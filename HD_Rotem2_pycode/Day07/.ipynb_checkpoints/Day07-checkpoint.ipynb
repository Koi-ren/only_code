```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver

url = "https://comic.naver.com/index"
driver = webdriver.Edge()
driver.get(url)
time.sleep(3)

soup = BeautifulSoup(driver.page_source, 'html.parser')

title = soup.select("li.AsideList__item--i30ly span.ContentTitle__title--e3qXt")
author = soup.select("li.AsideList__item--i30ly a.ContentAuthor__author--CTAAP")

driver.quit()

for i in range(5): print(f'순위: {i+1}, 제목: {title[i].text}, 작가: {author[i].text}')
```

```python
list1 = list([['허준호', '남자', 30, 183],
 ['이가원', '여자', 24, 162],
 ['배규민', '남자', 23, 179],
 ['고고림', '남자', 21, 182],
 ['이새봄', '여자', 28, 160],
 ['이보람', '여자', 26, 163],
 ['이루리', '여자', 24, 157],
 ['오다현', '여자', 24, 172]])
```

```python
list1
```

문자열로 처리하고 있다 ==> numpy의 ndarray로 처리 가능하다

자료형에 대해 알아야하는 것 ==> 각 자료마다 사용가능한 메소드가 다르기 때문임

```python
list1[1]
```

여자라는 성별을 알고싶다. list1[1,1]? 안된다 python의 list객체는 이러한 접근이 허용되지 않는다 ==> list[1][1]

```python
list1[1][1]
```

클래스가 다르다는 것은 객체에 대한 정의가 다르다는 뜻이기에... 클래스에 대한 정의가 올바른지 확인하는 것이 디버깅의 시작이다. 

기본은 입력과 출력에 대한 것에 정의가 잘 되어야 한다.
