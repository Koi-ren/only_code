```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver

url = "https://comic.naver.com/index"
driver = webdriver.Edge()
driver.get(url)
time.sleep(3)

soup = BeautifulSoup(driver.page_source, 'html.parser')

title = soup.select("li.AsideList__item--i30ly span.ContentTitle__title--e3qXt")
author = soup.select("li.AsideList__item--i30ly a.ContentAuthor__author--CTAAP")

driver.quit()

for i in range(5): print(f'순위: {i+1}, 제목: {title[i].text}, 작가: {author[i].text}')
```
