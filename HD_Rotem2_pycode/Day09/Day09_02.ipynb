```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/ws/HD_Rotem2_pycode/Day09/1_12_Seoul_tem.csv', encoding = 'cp949')
df.head(3)
```

```python
df.info()
```

```python
df2 = df.fillna(method='ffill')
df2.info()
```

```python
df2.head(3)
```

```python
plt.rcParams['font.family'] = 'gulim'
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.title('서울시 기온변화')
plt.plot(range(1,len(df2)+1), df2['최고기온(°C)'], label='최고기온(°C)', c='r') 
plt.plot(range(1,len(df2)+1), df2['평균기온(°C)'], label='평균기온(°C)', c='y')
plt.plot(range(1,len(df2)+1), df2['최저기온(°C)'], label='최저기온(°C)', c='b')
plt.xlabel('일')
plt.ylabel('기온')
plt.legend()
plt.show()
```
