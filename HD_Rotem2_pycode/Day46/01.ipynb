```python
from nltk import word_tokenize, sent_tokenize

#여러개의 문장으로 된 입력 데이터를 문장별로 단어를 토큰화(word 단위)
def tokenize_text(text):
    sentences = sent_tokenize(text)
    word_tokens = [word_tokenize(sent) for sent in sentences]
    return (word_tokens)

print(tokenize_text(text_sample))
```

```python
import nltk
nltk.download('stopwords')
```

```python
print('영어 stop words 갯수:',len(nltk.corpus.stopwords.words('english')))
print(nltk.corpus.stopwords.words('english')[:20])
```

```python
import re

stopwords = nltk.corpus.stopwords.words('english')
all_tokens = []

word_tokens = tokenize_text(text_sample)
for sentence in word_tokens: #한 문장에 대한 문장정보
    filtered_words=[]

    for word in sentence: #한 문장에서 word 읽기
        word = word.lower()

        if word not in stopwords:

            #re.sub은 대체하거나 제외 시키는 함수임
            word = re.sub(r"[^\w\s]","",word) # ^: 제외한, \w: 알파벳+숫자, \s: 공백(개행)
            # word = re.sub(r'[!"#$%&\'()*+,-./:;<=>?@\[\]^_\`{|}~\\\\ ]','', word)  #문장부호 없기기
            filtered_words.append(word)
    all_tokens.append(filtered_words)

print(all_tokens)
```
