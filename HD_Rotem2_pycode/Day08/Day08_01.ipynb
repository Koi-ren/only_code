따라서 질적 데이터를 볼때에는 또 다른 것을 사용해야하고, 이는 프로그래밍도 마찬가지이다. 

통계학, 코딩도 데이터 형식 -질적데이터, 양적데이터, 문자열, 정수형 등- 에 따른 문제해결방식(메소드)가 다르다.

분석을 할 떄에는 coloum단위의 데이터 조회를 하게 된다.

```python
def test(name, gender, age, height):
    print('이름: '+ name, end=", ")
    print('성별: ' + gender, end=", ")
    print('나이: ',age, end=", ")
    print('키: ',height)
    

test('홍길동', '남자', 20, 180)
```

```python
test('홍길동','남자',height=180,age=20)
```

개념만 이해하면, 객체지향으로의 입문은 쉽다 ==> 철학적 의미가 중요하기 때문이다

또한 함수 내에서 디폴트 값을 지정할 수 있다 ==> 또한 디폴트 값을 지정할 때엔 항상 뒤에서 부터 지정해줘야한다.

```python
def test(name, gender, height, age=20):
    print('이름: '+ name, end=", ")
    print('성별: ' + gender, end=", ")
    print('나이: ',age, end=", ")
    print('키: ',height)
```

```python
test('홍길동', '남성', 180)
```

함수를 사용하고 디폴트값을 변경해줄 때에도 뒤에서 부터 지정해줘야한다.  

```python
test('홍길동', '남성', 180, age=30)
```
