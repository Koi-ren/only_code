###  아래 과정에서 route에 의해 '/'뒤의 링크 생성, 

#### 아 셀을 실행 후...
#### ex) 
* <b> http://localhost:5000/ ---> 기본 링크
* http://localhost:5000/info
* http://localhost:5000/sec

<b>실행중인 셀을 강제 종료하는 방법: <br>실행중인 셀을 선택한 상태에서 i를 두번 클릭하면 interrupt됨

<b>http://localhost:5000/sec?name=박현호&age=25&gender=남
* sec_post에서 method를 post로 지정했는데 이때 GET에서는 args로 사용했던 것을 form으로 바꿔줘야한다.

```python
from flask import Flask, request

app = Flask(__name__)

# @로 어노테이션 실행
@app.route('/', methods=['GET'])
def home():
    return 'hello world'

@app.route('/info', methods=['GET'])
def info():
    title = "안녕하세요?    "
    return f'''
    <html>
    <head>
        <title>INFO</title>
    </head>
    <body>
        <h1>{title} 플라스크 서버정보</h1>
        <a href="/sec?title=링크연결">{title} 두번째 페이지</a>
    </body>
    </html>
    '''

@app.route('/sec', methods=['GET'])
def sec():
    title = request.args.get("title")
    name = request.args.get("name", "")
    age = request.args.get("age", "")
    gender = request.args.get("gender", "")
    return f'''
    <html>
    <head>
        <title>INFO</title>
    </head>
    <body>
        <h1>{title} 두번째 페이지</h1>
        <form method='post' action='/secpost'>
            <h3>이름: <input type="text" name="name" value="{name}"></h3>
            <h3>나이: <input type="text" name="age" value="{age}"></h3>
            <h3>성별: <input type="text" name="gender" value="{gender}"></h3>
            <button type='submit'>전송</button>
        </form>
    </body>
    </html>
    '''

@app.route('/secpost', methods=['POST'])
def sec_post():
    name = request.form.get("name")
    age = request.form.get("age")
    gender = request.form.get("gender")
    return f'''
    <html>
    <head>
        <title>INFO</title>
        <script>
            alert('{name}님 안녕하세요')
        </script>
    </head>
    <body>
        <h1>세번째 페이지</h1>
        <h3>이름: {name}</h3>
        <h3>나이: {age}</h3>
        <h3>성별: {gender}</h3>
    </body>
    </html>
    '''

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
