```python
from sklearn import svm, metrics
import glob, os.path, re, json, joblib
```

```python
def check_freq(fname):
    name = os.path.basename(fname)
    lang = re.match(r'^[a-z]{2,}', name).group()

    with open(fname, 'r', encoding="utf-8") as f:
        text = f.read()
        text = text.lower()

    cnt = [0 for n in range(0,26)]
    code_a = ord("a") #a의 아스키코드 숫자값 = 97
    code_z = ord("z") #z = 122

    for ch in text:
        n = ord(ch)
        if code_a <= n <= code_z:
            cnt[n - code_a] += 1

    ##정규화 
    total = sum(cnt)
    freq = list(map(lambda n: n/total , cnt))
    #명령어는 함수 형태로(lambda) 전환해줘야한다 
    
    return (freq, lang)
```

```python
# 경로 상에 있는 파일들 가져오기
def load_files(path):
    file_list = glob.glob(path)
    # print(file_list)
    freqs = [] #20개의 2차원 데이터
    labels = [] # 1차원 데이터
    
    for fname in file_list:
        # print(fname)
        # 하나의 파일을 정형구조로 생성할것
        r = check_freq(fname) # 빈도수를 체크하는 기능 -> 이 내용에서 구현하면 길어지고 복잡해보이니 따로 함수로 만들었어
        freqs.append(r[0]) #2차원
        labels.append(r[1]) #1차원
        
    return {"freqs": freqs, "labels": labels}
```

```python
data = load_files("c:/ws/HD_Rotem2_pycode/Day31_언어판별/lang_data/train/*.txt")
test = load_files("c:/ws/HD_Rotem2_pycode/Day31_언어판별/lang_data/test/*.txt")

with open("c:/ws/HD_Rotem2_pycode/Day31_언어판별/lang_data/freq2.json", "w", encoding="utf-8") as f:
    json.dump([data,test], f)
#data['lebels']
```

```python
clf = svm.SVC()
clf.fit(data["freqs"], data["labels"])

# 예측하기 --- (※5)
#여기서는 적용하지 않았으나 원핫인코딩 방식 적용하는 것을 추천
predict = clf.predict(test["freqs"])

# 결과 테스트하기 --- (※6)
ac_score = metrics.accuracy_score(test["labels"], predict)
cl_report = metrics.classification_report(test["labels"], predict)
print("정답률 =", ac_score)
print("리포트 =")
print(cl_report)
```

```python
joblib.dump(clf, "c:/ws/HD_Rotem2_pycode/Day31_언어판별/lang_data/freq2.pkl")
print("ok")
```
