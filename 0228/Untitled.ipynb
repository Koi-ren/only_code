```python
import os
import shutil
import random

def split_and_copy_images(src_folders, dest_root="dataset", split_ratio=0.8):
    random.seed(42)  # 재현 가능성을 위해 시드 설정

    for category in src_folders:
        src_path = os.path.join("img_data", category)
        train_path = os.path.join(dest_root, "train", category)
        test_path = os.path.join(dest_root, "test", category)

        os.makedirs(train_path, exist_ok=True)
        os.makedirs(test_path, exist_ok=True)

        images = [f for f in os.listdir(src_path) if f.lower().endswith(".jpg")]
        random.shuffle(images)

        split_idx = int(len(images) * split_ratio)
        train_files = images[:split_idx]
        test_files = images[split_idx:]

        for i, file in enumerate(train_files):
            new_name = f"{category}_{i}.jpg"
            shutil.copy2(os.path.join(src_path, file), os.path.join(train_path, new_name))

        for i, file in enumerate(test_files):
            new_name = f"{category}_{i + len(train_files)}.jpg"
            shutil.copy2(os.path.join(src_path, file), os.path.join(test_path, new_name))

        print(f"Category: {category} - Train: {len(train_files)}, Test: {len(test_files)}")

# 실행
data_categories = ["pikachu", "hachuping"]
split_and_copy_images(data_categories)
```
