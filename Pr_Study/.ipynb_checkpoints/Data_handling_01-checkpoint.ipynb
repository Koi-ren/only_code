```python
import numpy as np
```

넘파이의 기반 데이터 타입은 ndarray <br>
넘파이에서 다차원(Multi-dimension)배열을 생성, 연산 수행 가능

```python
array1 = np.array([1, 2, 3])
print('array1 type:', type(array1))
print('array1 array 형태: ', array1.shape)
```

```python
array2 = np.array([[1, 2, 3],[2, 3, 4]])
print('array2 type:', type(array2))
print('array2 array 형태: ', array2.shape)
```

```python
array3 = np.array([[1, 2, 3]])
print('array3 type:', type(array3))
print('array3 array 형태: ', array3.shape)
```

```python
print('array1: {0}차원, array2: {1}차원, array3: {2}차원'. format(array1.ndim, 
                                                            array2.ndim, array3.ndim))
```

```python
list1 = [1, 2, 3]
print(type(list1))

array1 = np.array(list1)
print(type(array1))
print(array1, array1.dtype)
```

위와 같이 리스트에 nd.array()함수를 사용한다면 ndarray로 변환된 list 속 데이터 형태는 모두 int32형으로 통일됨. <br>
이는 ndarray 특성으로 같은 모든 데이터 타입이 특정 형태로로 변환됨.

```python
list2 = [1, 2, 'test']
array2 = np.array(list2)
print(array2, array2.dtype)
list3 = (1, 2, 3.0)
array3 = np.array(list3)
print(array3, array3.dtype)
```
